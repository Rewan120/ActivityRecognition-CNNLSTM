```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
```

```python
# 1. تحديد المسارات
train_images_dir = '/content/drive/MyDrive/archive (1)/Human Action Recognition/train'  # مسار الصور
train_labels_file = '/content/drive/MyDrive/archive (1)/Human Action Recognition/Training_set.csv'  # ملف الفئات

# 2. تحميل البيانات من ملف CSV
train_data = pd.read_csv(train_labels_file)

# عرض أول 5 صفوف من البيانات للتأكد
print(train_data.head())
```

```python
# 3. تحويل الصور إلى بيانات قابلة للتدريب
X_train = []
y_train = []

for filename, label in zip(train_data['filename'], train_data['label']):
    img_path = os.path.join(train_images_dir, filename)
    try:
        img = load_img(img_path, target_size=(64, 64))  # تعديل الحجم حسب الحاجة
        img_array = img_to_array(img)
        X_train.append(img_array)
        y_train.append(label)
    except FileNotFoundError:
        print(f"الصورة {filename} غير موجودة.")

# تحويل البيانات إلى numpy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)

print(f"عدد الصور التي تم تحميلها: {len(X_train)}")
```

```python
# 4. ترميز الفئات إلى قيم رقمية باستخدام LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# تحويل الأهداف إلى شكل One-Hot Encoding
y_train_encoded = to_categorical(y_train_encoded)

print(f"عدد الفئات الفريدة: {len(label_encoder.classes_)}")
```

```python
# 5. تقسيم البيانات إلى مجموعة تدريب واختبار
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X_train, y_train_encoded, test_size=0.2, random_state=42)

print(f"تم تقسيم البيانات بنجاح: {len(X_train_split)} صورة للتدريب, {len(X_test_split)} صورة للاختبار.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, LSTM, Dense, Dropout, Reshape

# بناء نموذج CNN-LSTM
model = Sequential()

# إضافة طبقة Conv2D لاستخراج الملامح من الصور
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))  # 64x64 صورة مع 3 قنوات لونية (RGB)
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# إعادة تشكيل الأبعاد لتناسب LSTM
# تحويل البيانات من (batch_size, 6, 6, 128) إلى (batch_size, 36, 128)
model.add(Reshape((6 * 6, 128)))  # 6*6 هو عدد التوقيتات و128 هو عدد الخصائص

# إضافة طبقة LSTM
model.add(LSTM(64, return_sequences=False))

# إضافة طبقة Dense للتصنيف
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))  # لمنع الإفراط في التكيف
model.add(Dense(5, activation='softmax'))  # 5 فئات، يمكنك تعديلها حسب عدد الفئات في البيانات الخاصة بك

# تجميع النموذج باستخدام خوارزمية Adam
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# عرض ملخص النموذج
model.summary()
```

```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# تحويل البيانات إلى One-Hot
y_train_split = to_categorical(y_train_split, num_classes=15)
y_test_split = to_categorical(y_test_split, num_classes=15)

# تعديل النموذج ليشمل تحسينات
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(64, 64, 3)),  # تأكد من أن input_shape صحيح
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(15, activation='softmax')  # عدد الفئات 15
])

# تجميع النموذج مع `categorical_crossentropy`
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# إعداد EarlyStopping و ReduceLROnPlateau
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3)

# تدريب النموذج مع إضافة الـ Callbacks
history = model.fit(
    X_train_split, y_train_split,
    epochs=10,
    batch_size=32,
    validation_data=(X_test_split, y_test_split),
    callbacks=[early_stopping, reduce_lr]
)
```

```python
from tensorflow.keras.layers import GlobalAveragePooling2D

# نموذج محسن
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(64, 64, 3)),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.GlobalAveragePooling2D(),  # تحسين الطبقة
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(15, activation='softmax')
])

# تجميع النموذج مع تحسينات
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3)

# تدريب النموذج
history = model.fit(
    X_train_split, y_train_split,
    epochs=10,
    batch_size=32,
    validation_data=(X_test_split, y_test_split),
    callbacks=[early_stopping, reduce_lr]
)
```
